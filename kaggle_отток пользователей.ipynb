Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)

# Применяем one-hot encoder к категориальным признакам
one_hot_encoded = encoder.fit_transform(data[cat_cols])
one_hot_encoded_test = encoder.fit_transform(test_data[cat_cols])

# Делаем датафрейм из преобразованных колонок
one_hot_data = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cat_cols), index=data.index)
one_hot_test_data = pd.DataFrame(one_hot_encoded_test, columns=encoder.get_feature_names_out(cat_cols), index=test_data.index)
```

```python
# Соединеям обработанные категориальные и численные признаки

data_encoded = pd.concat([data, one_hot_data], axis=1)
test_data_encoded = pd.concat([test_data, one_hot_test_data], axis=1)
data_encoded = data_encoded.drop(cat_cols, axis=1)
test_data_encoded = test_data_encoded.drop(cat_cols, axis=1)
```

```python
X = data_encoded.drop([target_col], axis=1)
y = data_encoded[target_col]
X_test = test_data_encoded
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.2, random_state=42)
```

```python
# Нормируем числовые признаки

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegressionCV
# Применяем LogisticRegression

linear_model = LogisticRegressionCV(Cs=[0.001, 0.01, 0.1, 1, 10, 100], scoring='roc_auc', refit=True)
linear_model.fit(X_train_scaled, y_train)
y_train_predicted = linear_model.predict_proba(X_train_scaled)[:, 1]
y_valid_predicted = linear_model.predict_proba(X_valid_scaled)[:, 1]
y_pred = linear_model.predict_proba(X_test_scaled)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
valid_auc = roc_auc_score(y_valid, y_valid_predicted)
```
